```python
path = './data1/' # data path

a = 10.  # simulation length (-a,a)
uk =1.   #
k = np.pi / a * uk
num = 1e5 # particle number
m=1.
tem = 1 # initial temperature
kappa =0 #initial kappa index, kappa=0 means Maxwellian

q = np.sqrt(2.*a/num)
dens = num / a / 2.
lambD2 = tem / dens  # $\lambda_D^2$
plasmaFreq2 = dens

gridNum = 200  # grid number
gridW = 2 * a / gridNum #grid width 

xs = np.linspace(-a, a, num=gridNum, endpoint=False)
dxs = xs[1]-xs[0]

vmax = 10* np.sqrt(tem) # max velocity in simulation
vs = np.linspace(-vmax, vmax, num=2000, endpoint=False)

tmax = 500000  # max time step
t = np.linspace(0, tmax, tmax)
dt = 0.1 # time step

uae=0.6 # initial electrons amplitude
uai=0.4 #ion amplitude
```

## Definition of distribution function
- 1D Maxwellian velocity distribution：fm(v,t)
$$ f_M(v) = \sqrt{ \frac{1}{2 \pi k T} } \exp(- \frac{m v^2}{2kT}) $$ 
- 1D kappa velocity distribution：fk(v,t,kappa)
$$ f_{\kappa}(v) = \sqrt{ \frac{1}{2 \pi k T (\kappa-1/2)} } \frac{\Gamma(\kappa+1)}{\Gamma(\kappa+1/2)} ( 1+\frac{m v^2}{(\kappa-1/2) 2kT} )^{-\kappa-1}$$
- initial density distribution：ue/ui
$$ u_e(x) = 1+uae\cos(\frac{\pi x}{a})$$
$$ u_i(x) = 1+uai\cos(\frac{\pi x}{a})$$

```python
def fm(v, t):
    r = np.sqrt(m/(2.*np.pi*t))*np.exp(-m*v**2/(2*t))
    return r
def fk(v, t, kappa):
    if(kappa == 0):
        r = fm(v, t)
    else:
        r = np.sqrt(m/(2.*np.pi*t*kappa)) * sp.gamma(kappa+1.5) / \
            sp.gamma(kappa+1) * (1 + m*v**2 / kappa / 2 / t) ** (-kappa-1.5)
    return r
def ue(x):
    r = (1. + uae * np.cos(k*x))
    return r
def ui(x):
    r = (1. + uai * np.cos(k*x))
    return r
```

### Solving the parameters $\kappa_0$ and $T_0$

#### calculating the VDF of simulation

```python
def pdf(data, dmin, dmax):# bins = 20
    bins_optional = np.histogram_bin_edges(
        data, bins=20, range=(dmin, dmax))
    pdf_v = np.histogram(data, bins=bins_optional, density=True)[0]
    pdfr = (bins_optional[0:len(bins_optional)-1]+bins_optional[1:len(bins_optional)])/2
    pdfr_w = bins_optional[1:len(bins_optional)]-bins_optional[0:len(bins_optional)-1]
    return pdf_v, pdfr, pdfr_w
def pdf2d(x,v,xmin,xmax,vmin,vmax):
    x_bins = np.histogram_bin_edges(x, bins='auto', range=(xmin, xmax))
    v_bins = np.histogram_bin_edges(v, bins='auto', range=(vmin, vmax))
    pdf = np.histogram2d(x,v,bins=[x_bins,v_bins],density=True)[0]
    pdfr_x = (x_bins[0:len(x_bins)-1]+x_bins[1:len(x_bins)])/2
    pdfr_v = (v_bins[0:len(v_bins)-1]+v_bins[1:len(v_bins)])/2
    pdfr_xw = x_bins[1:len(x_bins)]-x_bins[0:len(x_bins)-1]
    pdfr_vw = v_bins[1:len(v_bins)]-v_bins[0:len(v_bins)-1]
    return pdf,pdfr_x,pdfr_v,pdfr_xw,pdfr_vw

xvipdf,xib,vib,xibw,vibw= pdf2d(xidata,vxidata,-a,a,-vmax,vmax)
xvpdf,xb,vb,xbw,vbw = pdf2d(xdata,vxdata,-a,a,-vmax,vmax)

#initial VDF
vipdf, vir, virw=pdf(vxidata,-vmax,vmax)
#final VDF
vfpdf, vfr, vfrw=pdf(vxdata,-vmax,vmax)
```

#### Verification of conservations

- energy conservation

```python
# simulation results
ekinit = np.sum(datani*datati)/num*gridW*0.5
epinit = np.sum(dataci*datafi)/num*gridW*0.5
einit = ekinit+epinit

efin = (np.sum(datan*datat)+np.sum(datac*dataf))/num*gridW*0.5
ekfin = (np.sum(datan*datat))/num*gridW*0.5
epfin = (np.sum(datac*dataf))/num*gridW*0.5

#theoretical energy
def nphi(x):
    phi = -( dens*q*(uae-uai) )/ (k**2)*np.cos(k*x)/2./a
    return (ui(x)-ue(x) )*q*phi
ek_theo = 0.5*tem
ep_theo = 0.5*integrate.quad(nphi,-a,a)[0]
einit_theo=ek_theo+ep_theo

energy_table = PrettyTable(['','total energy','kinetic energy','potential energy'])
energy_table.add_row(['simulation(initial)',einit,ekinit,epinit])
energy_table.add_row(['simulation(final)',efin,ekfin,epfin])
energy_table.add_row(['theory',einit_theo,ek_theo,ep_theo])

print(energy_table)
```

- entropy conservation

```python
def e(pdf,xw,vw):
    r = 0.0
    for x in pdf:
        for i in x:
            if i == 0.0:
                r += 0.0
            else: 
                r+=- i * np.log(i)
    return r * vw * xw

xi_e = np.linspace(-a,a,num=2000,endpoint=False)
vi_e = np.linspace(-100*vmax, 100*vmax, num=20000, endpoint=False)

def entropy(x,v):
    ufv = fk(v, tem,kappa)*ue(x)/2./a
    if (ufv==0.0):
        r = 0.0
    else:
        r = -ufv * np.log(ufv)
    return r
options = {'limit':50000}
entropyt_int_res = integrate.nquad(entropy,[[-a,a],[-np.inf,np.inf]], opts=[options,options])

entropyi=e(xvipdf,xibw[0],vibw[0])
entropyf=e(xvpdf,xbw[0],vbw[0])
entropyt=entropyt_int_res[0]

entropy_table = PrettyTable(['','entropy'])
entropy_table.add_row(['simulation(intial)',entropyi])
entropy_table.add_row(['simulation(final)',entropyf])
entropy_table.add_row(['theory',entropyt])

print(entropy_table)
```

#### Solving the following equations with perodic boundary condition
- the equation of ue(x)
$$ U'' - \frac{\kappa'_0+2}{\kappa'_0+1}\frac{(U')^2}{U} - \frac{\kappa'_0+1}{\kappa'_0} \frac{n_0 e^2}{\varepsilon_0 k T'_0} U^{\frac{1}{\kappa'_0+1}+1} (U-u) =0 $$
- energy equation
$$ \iint n_e k T dx = \iint n'_e k T'dx - \iint (n'_e-n'_i) e \varphi' dx$$
- entropy equation
$$ \iint f \ln f dx dv = \iint F \ln F dx dv$$


```python
def ug(x):
    r = (1. + uai * np.cos(k*x)) #/ (2*1.*a)
    return r

def ug_dev(x):
    r = -uai*k*np.sin(k*x)#/ 2. / a
    return r

def eq4U(x,yy,p):
    kappa,t=p
    dy, y = yy
    k_factor = (kappa+2) / (kappa+1)
    dy2y = (dy**2) / y
    ddy = k_factor * dy2y + (kappa+1) / (kappa) * (y**k_factor) * ( y-ui(x) ) / t * dens * (q**2)
    return np.vstack((ddy, dy))

def bc4U(ya, yb):
    return np.array([ya[0]-yb[0],ya[1]-yb[1]])

xi = np.linspace(-a, a, num=5000, dtype='complex',endpoint=False)
yguess = np.array([ug_dev(xi), ui(xi)])

def eq4para(x,energy,entropy):
    kp,tp = x
    sol4ode = solve_bvp(partial(eq4U, p=x), bc4U, xi,
                        yguess, tol=1e-13, max_nodes=50000)

    def ff(x):
        return sol4ode.sol(x)[1]
    def ff_r(x):
        return ff(x).real
    def ff_i(x):
        return ff(x).imag

    def ff_inti(x):
        r = ff(x)**(kp/(kp+1))
        return r
    def ff_inti_r(x):
        return ff_inti(x).real
    def ff_inti_i(x):
        return ff_inti(x).imag
        
    def ff_intii(x):
        r = ui(x) * (ff(x) ** (-1.0/(kp+1)))
        return r
    def ff_intii_r(x):
        return ff_intii(x).real
    def ff_intii_i(x):
        return ff_intii(x).imag
    
    def ff_intiii(x):
        r = ff(x)**(-1./(kp+1))
        return r
    def ff_intiii_r(x):
        return ff_intiii(x).real
    def ff_intiii_i(x):
        return ff_intiii(x).imag
    
    def entropy_pred(x, v):
        U_x = ff(x)
        
        if(kp==0):
            t_x = tp
            tempf = U_x * fm(v, t_x)+0j
        else:
            t_x = tp * U_x ** (-1./(kp+1))
            tempf = U_x * fk(v, t_x, kp)/2./a+0j
        if (tempf == 0.0+0.0j):
            r = 0.0
        else:
            r = -tempf * np.log(tempf)
        r = -tempf * np.log(tempf)
        return r
    def entropy_pred_r(x,v):
        return entropy_pred(x,v).real
    def entropy_pred_i(x,v):
        return entropy_pred(x,v).imag
    entropy_pred_int_r = integrate.dblquad(
        entropy_pred_r, -np.inf, np.inf,  -a,  a, epsrel=1e-10)
    entropy_pred_int_i = integrate.dblquad(
        entropy_pred_i, -np.inf, np.inf,  -a,  a, epsrel=1e-10)
    entropyp = entropy_pred_int_r[0] + 1j * entropy_pred_int_i[0]

    int_ff_r = integrate.quad(ff_r, -a, a, epsrel=1e-10)[0]-1.0
    int_ff_i = integrate.quad(ff_i, -a, a, epsrel=1e-10)[0]-1.0
    int_ff = int_ff_r + 1j * int_ff_i
    
    int_i_r = integrate.quad(ff_inti_r, -a, a, epsrel=1e-10)[0]
    int_i_i = integrate.quad(ff_inti_i, -a, a, epsrel=1e-10)[0]
    int_i = int_i_r + 1j * int_i_i

    int_ii_r = integrate.quad(ff_intii_r, -a, a, epsrel=1e-10)[0]
    int_ii_i = integrate.quad(ff_intii_i, -a, a, epsrel=1e-10)[0]
    int_ii = int_ii_r + 1j * int_ii_i

    int_iii_r = integrate.quad(ff_intiii_r, -a, a, epsrel=1e-10)[0]
    int_iii_i = integrate.quad(ff_intiii_i, -a, a, epsrel=1e-10)[0] 
    int_iii = (int_iii_r + 1j * int_iii_i) * int_ff

    eq1 = ((energy*2)*2*a - tp * (1+kp) * int_i + tp * kp * int_ii)/(energy*2)/2./a
    eq2 = (entropyp-entropy)/entropy

    return eq1.real,eq2.real
```

```python
para = [0.14708099,2.83689694]# initial guess of parameters

sol_para_it = op.root(partial(eq4para,energy=einit_theo,entropy=entropyt), para, method='hybr')
sol_para_fs = op.root(partial(eq4para,energy=efin,entropy=entropyf), para, method='hybr')

parait = sol_para_it.x
parafs = sol_para_fs.x

print(parait)
print(sol_para_it.fun,sol_para_it.success)
print(parafs)
print(sol_para_fs.fun,sol_para_fs.success)
```

### Velocity distribution
$$ f_{\kappa}(x,v) = U(x) \sqrt{ \frac{U(x)^{ \frac{1}{\kappa'_0+1} }}{2 \pi k T'_0 \kappa'_0} } \frac{\Gamma(\kappa'_0+3/2)}{\Gamma(\kappa'_0+1)} ( 1+\frac{m v^2}{\kappa'_0 2kT'_0}*U(x)^{ \frac{1}{\kappa'_0+1} } )^{-\kappa-1}$$

#### Global velocity distribution

```python
def fk_xv(x,v,k,t):
    if(k==0):
        s = ff(x)*np.sqrt( ff(x)**( 1./(k+1) ) / (2*np.pi*t) )/2./a
        m = np.exp(-v**2 / (2*t) * ff(x)**(1./(k+1)))
        r = s*m
    else:
        s = ff(x)*np.sqrt( ff(x)**( 1./(k+1) ) / (2*np.pi*t*k) )/2./a
        frac = sp.gamma(k+1.5) / sp.gamma(k+1)
        m = ( 1+ (v)**2 / (k*2*t) * ff(x)**(1./(k+1)) )**(-k-1.5)
        r = s*frac*m
    return r.real
def fk_v_nonvec(vv,kk,tt):
    r = integrate.quad(partial(fk_xv,v=vv,k=kk,t=tt),-a,a)[0]
    return r
fk_v = np.vectorize(fk_v_nonvec)
ax1 = plt.gca()

ax1.plot(vfr, vfpdf, '.', color='r', label='simulation')
ax1.plot(vs, fk_v(vs, parafs[0],parafs[1]),'--',color='blue',label='theory')
ax1.plot(vs, fk(vs, tem,kappa),color='orange',label='initial PDF')
ax1.set_xlabel(r'$v$')
ax1.set_ylabel(r'$f_{v,e}(v)$')

ax1.legend()
ax1.grid()
ax1.set_yscale('log')
#plt.savefig('./figs/vpdf2.png',dpi=600)
plt.savefig('./figs/vpdf1.eps',bbox_inches = 'tight')
plt.show()
chi_err_f = np.sum(
    ( (fk_v(vfr, parafs[0], parafs[1])-vfpdf) )**2/fk_v(vfr, parafs[0], parafs[1]) * vfrw)
print(chi_err_f)
```

```python
def fkit_xv(x,v,k,t):
    if(k==0):
        s = ffit(x)*np.sqrt( ffit(x)**( 1./(k+1) ) / (2*np.pi*t) )/2./a
        m = np.exp(-v**2 / (2*t) * ffit(x)**(1./(k+1)))
        r = s*m
    else:
        s = ffit(x)*np.sqrt( ffit(x)**( 1./(k+1) ) / (2*np.pi*t*k) )/2./a
        frac = sp.gamma(k+1.5) / sp.gamma(k+1)
        m = ( 1+ (v)**2 / (k*2*t) * ffit(x)**(1./(k+1)) )**(-k-1.5)
        r = s*frac*m
    return r.real
def fkit_v_nonvec(vv,kk,tt):
    r = integrate.quad(partial(fkit_xv,v=vv,k=kk,t=tt),-a,a)[0]
    return r
fkit_v = np.vectorize(fkit_v_nonvec)
def fkfs_xv(x,v,k,t):
    if(k==0):
        s = fffs(x)*np.sqrt( fffs(x)**( 1./(k+1) ) / (2*np.pi*t) )/2./a
        m = np.exp(-v**2 / (2*t) * fffs(x)**(1./(k+1)))
        r = s*m
    else:
        s = fffs(x)*np.sqrt( fffs(x)**( 1./(k+1) ) / (2*np.pi*t*k) )/2./a
        frac = sp.gamma(k+1.5) / sp.gamma(k+1)
        m = ( 1+ (v)**2 / (k*2*t) * fffs(x)**(1./(k+1)) )**(-k-1.5)
        r = s*frac*m
    return r.real
def fkfs_v_nonvec(vv,kk,tt):
    r = integrate.quad(partial(fkfs_xv,v=vv,k=kk,t=tt),-a,a)[0]
    return r
fkfs_v = np.vectorize(fkfs_v_nonvec)
ax1 = plt.gca()

ax1.plot(vfr, vfpdf, '.', color='r', label='simulation')
ax1.plot(vs, fkit_v(vs, parait[0],parait[1]),'--',color='blue',label='theory')
ax1.plot(vs, fkfs_v(vs, parafs[0],parafs[1]),'-',color='black',label='semi-theory')

ax1.set_xlabel(r'$v$')
ax1.set_ylabel(r'$f_{v,e}(v)$')

ax1.legend()
ax1.grid()
ax1.set_yscale('log')
#plt.savefig('./figs/com-vpdf.png',dpi=600)
plt.savefig('./figs/com-vpdf.eps',bbox_inches = 'tight')
plt.show()
```
